```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset/ecommerce.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['Address', 'Email'], axis=1, inplace=True)
```

```python
df
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_input = df[['Avg. Session Length', 'Time on App', 'Time on Website',
       'Length of Membership']]
```

```python
data_output = df['Yearly Amount Spent']
```

```python
data_input
```

```python
data_output
```

```python
df.info()
```

```python
df.corr()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.25, random_state=50)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_, index=['Avg. Session Length', 'Time on App', 'Time on Website','Length of Membership'], columns=['coef'])
```
